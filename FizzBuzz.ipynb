### Createing a Fizz Buzz Project

If a number is divisor of three (3) and five (5) print FizzBuzz

If a number is a divisor of three (3) print Fizz

If a number is a divisor of five (5) print Buzz

```python
# Create a fizzbuzz
n = int(input("Enter a number: "))

# Iterate through over the number that has been entered
for index, i in enumerate(range(n)):

    # Check through to determine if the range of numbers is a fizz or buzz
    if i % 3 == 0 and i % 5 == 0:
        print(index, "--> FizzBuzz")
    elif i % 3 == 0:
        print(index, "--> Fizz")
    elif i % 5 == 0:
        print(index, "--> Buzz")
    else:
        print(index, i)
    
```
